```python
train=pd.read_csv('http://bit.ly/kaggletrain')
train.head()
```

```python
train['Sex_male']=train.Sex.map({'female':0,'male':1})
train.head()
#sex_male diye bir sütun oluşturduk, buna değerleri Sex sütunundaki female olanlara 0
#male olanlara 1 atayan bir dictionary oluşturarak yaptık.
```

```python
#Dummy oluştururken K tane değer varsa K-1 sütun yeterli oluyor.

pd.get_dummies(train.Sex).iloc[:,1:].head()
#Şeklinde Sex sütununda bütün satırları alıyoruz sütunlara 0 dan degil 1 den başlatarak ilk sütunu almıyoruz. female sütununu almamıs oluyoruz.
```

```python
pd.get_dummies(train.Sex,prefix='Sex').iloc[:,1:].head()
```

```python
pd.get_dummies(train.Embarked,prefix='Embarked').head(10)
```

```python
pd.get_dummies(train.Embarked,prefix='Embarked').iloc[:,1:].head(10)
```

```python
embarked_dummies=pd.get_dummies(train.Embarked,prefix='Embarked').iloc[:,1:]
train=pd.concat([train,embarked_dummies],axis=1)
train.head()
#oluşturdugumuz dummyleri degiskene atadik.
#bu değişken ile train veri setini sütunlar(axis=1) boyunca birleştirdik.
```
